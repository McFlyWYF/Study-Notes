## 偏函数

* Python的functools模块其中一个功能就是偏函数，通过设定参数的默认值，降低参数调用的难度。

```python
#默认转换为十进制
int('123')
```

```python
#base参数可以做N进制转换
int('123',base=8)
```

```python
int('12',16)
```

```python
def int2(x,base=2):
    return int(x,base)
```

```python
int2('101')
```

```python
#functools.partial帮助我们曾创建一个偏函数，自己创建
import functools
int2 = functools.partial(int,base=2)
```

```python
int2('101')
```

```python
int2('101',base=10)
```

```python
#创建偏函数时，可以接收函数对象、*args和**kw参数
int2 = functools.partial(int,base=2)
```

```python
int('101')
```

相当于

```python
kw = {'base':2}
int('101',**kw)
```

当传入

```python
max2 = functools.partial(max,10)
```

```python
max2(1,2,3)
```

相当于

```python
args = (10,5,6,7)
max(*args)
```
