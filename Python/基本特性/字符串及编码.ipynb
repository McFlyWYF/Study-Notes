### 常量

* 一种除法是`/`：

```python
10 / 3
```

`/`除法计算结果是浮点数，即使是两个整数恰好整除，结果也是浮点数：

```python
9 / 3
```

* 另一种除法是`//`：

```python
10 // 3
```

```python
5 // 3
```

这种除法称为地板除，两个整数的除法仍然是整数，即使除不尽，永远是整数。

### 字符编码

* ASCII编码一个字节表示的最大整数是255，用一个字节表示一个字符。Unicode用两个字节表示一个字符。在计算机内存中，使用Unicode编码，当需要保存到硬盘或者传输的时候，就转换为UTF-8编码。

* 把bytes变为str，需要用decode()方法：

```python
b'ABC'.decode('ascii')
```

```python
b'\xe4\xb8\xad\xe6\x96\x87'.decode('utf-8')
```

* 计算str包含多少个字符，使用len函数：

```python
len('ABC')#计算字符数
```

```python
len(b'ABC')#转换为bytes，计算字节数
```

* 格式化字符串

```python
'Hello,%s' % 'world'
```

```python
'Hi,%s,you have $%d.' % ('Michael',1000000)
```

* 常用的占位符有：`%d`：整数,`%f`：浮点数,`%s`：字符串,`%x`：十六进制整数

* 格式化整数和浮点数还可以指定是否补0和整数与小数的位数：

```python
'%2d-%02d' % (3,1)
```

```python
'%.2f' % 3.1415926
```

* %s会把任何数据类型转换为字符串：

```python
'Age:%s.Gender:%s' % (25,True)
```

```python
'growth rate:%d %%' % 7#使用%%来表示一个%
```
