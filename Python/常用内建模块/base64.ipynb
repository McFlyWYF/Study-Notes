## base64

base64是一种用64个字符来表示任意二进制数据的方法。

* 首先准备一个包含64个字符的数组。['A','B',...,'a',..,'0',...,'+','/']
* 然后对二进制数据进行处理，3个字节一组，一共是 3x8=24bit，划为 4 组，每组正好 6 个 bit：
* 这样我们得到 4 个数字作为索引，然后查表，获得相应的 4 个字符，就
是编码后的字符串。

Base64 编码会把 3 字节的二进制数据编码为 4 字节的文本数据。

如果要编码的二进制数据不是 3 的倍数，最后会剩下 1 个或 2 个字节怎
么办？Base64 用\x00 字节在末尾补足后，再在编码的末尾加上 1 个或 2
个=号，表示补了多少字节，解码的时候，会自动去掉。

```python
import base64
base64.b64encode(b'binary\x00string')#编码
```

```python
base64.b64decode(b'YmluYXJ5AHN0cmluZw==')#解码
```

url safe把字符+和/分别变成`-`和`_`。

```python
base64.b64encode(b'i\xb7\x1d\xfb\xef\xff')
```

```python
base64.urlsafe_b64encode(b'i\xb7\x1d\xfb\xef\xff')
```

```python
base64.urlsafe_b64decode('abcd--__')
```
