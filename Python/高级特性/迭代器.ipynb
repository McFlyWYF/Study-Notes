## 迭代器

* 可以作用于for循环的数据类型有：
* 一类是list、dict、tuple、set、str等。
* 一类是generator，包括生成器和带yield的generator function|

### Iteratable：可迭代对象
### Iterator：迭代器

```python
#判断一个对象是否是Iterable对象
from collections import Iterable
isinstance([],Iterable)
```

```python
isinstance({},Iterable)
```

```python
isinstance('abc',Iterable)
```

```python
isinstance((x for x in range(10)),Iterable)
```

```python
isinstance(100,Iterable)
```

* 可以被next()函数调用并不断返回下一个值的对象称为迭代器，Iterator

```python
from collections import Iterator
#判断是否是Iterator对象
isinstance([],Iterator)
```

```python
isinstance({},Iterator)
```

```python
isinstance((x for x in range(10)),Iterator)
```

```python
isinstance('abc',Iterator)
```

生成器都是Iterator对象，但list，dict，str虽然是Iteratable，但不是Iterator。

```python
#把list，dict，str等Iterable变成Iterator可以使用iter()函数
isinstance(iter([]),Iterator)
```

```python
isinstance(iter('abc'),Iterator)
```

* 为什么list，str，dict等数据类型不是Iterator？

```python
for x in [1,2,3,4,5]:
    print(x)
```

```python
#等价于
it = iter([1,2,3,4,5])#获得Iterator对象
while True:
    try:
        x = next(it)#获得下一个值
        print(x)
    except StopIteration:#遇到异常退出循环
        break
```
