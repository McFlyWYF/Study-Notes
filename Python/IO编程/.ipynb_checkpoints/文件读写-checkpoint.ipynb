## 文件读写

### 读文件

open()函数

```python
#打开文件，不存在抛出错误
f = open('test.txt','r')
```

```python
#读取文件内容
f.read()
```

```python
#关闭文件
f.close()
```

文件读写时有可能产生IOError，一旦出错，f.close()不会执行。

```python
try:
    f = open('test.txt','r')
finally:
    if f:
        f.close()
```

使用with语句自动调用close()方法。

```python
with open('test.txt','r') as f:
    print(f.read())
```

* read()读取所以内容
* read(size)读取size字节的内容
* readline()读取一行内容
* readlines()读取所以内容，按行返回list

```python
f.read(2)
```

```python
f.readline()
```

```python
f.readline()
```

```python
f.readlines()
```

```python
f.read()
```

```python
for line in f.readlines():
    print(line.strip())#去掉末尾'\n'
```

### file-like Object

### 二进制文件

使用`rb`模式读取二进制文件

```python
f = open('James2.jpg','rb')
f.read()
```

### 字节编码

读取非UTF-8编码的文件，需要给open函数传入encoding参数。

```python
f = open('test.txt','r',encoding='gbk')
f.read()
```

open()函数还接收一个errors参数，表示如果遇到编码错误后如何处理。

```python
f = open('test.txt','r',encoding='gbk',errors='ignore')
```

```python
f.read()
```

### 写文件

```python
f = open('test.txt','w')
f.write('Hello,world!')
f.close()
```

```python
with open('test.txt','w',encoding='gbk') as f:
    f.write('Hello')
```
