## 继承和多态

```python
class Animal(object):
    def run(self):
        print('Animal is running...')
```

```python
#继承自Animal，可以得到父类的所以功能
class Dog(Animal):
    #对子类增加自己的方法
    def run(self):
        print('Dog is running...')
        
    def eat(self):
        print('Eating meat...')

class Cat(Animal):
    
    def run(self):
        print('Cat is running...')
```

```python
dog = Dog()
dog.run()#自动用于run方法
dog.eat()
```

```python
cat = Cat()
cat.run()
```

当子类和父类都存在相同的run方法时，子类的run方法覆盖了父类的run方法，调用子类的run方法。

```python
isinstance(cat,Animal)
```

```python
isinstance(cat,Cat)
```

```python
a = Animal()
isinstance(a,Animal)
```

```python
#不能说Animal是Dog的对象
isinstance(a,Dog)
```

### 多态

```python
def run_twice(animal):
    animal.run()
    animal.run()
```

```python
#传入Animal对象，调用Animal的run方法
run_twice(Animal())
```

```python
#传入Dog对象，调用Dog的run方法
run_twice(Dog())
```

```python
class Tortoise(Animal):
    def run(self):
        print('Tortoise is running slowly...')
```

```python
run_twice(Tortoise())
```

多态的好处就是，当我们需要传入Dog、Cat、Tortoise时，我们只需要接收Animal类型就可以了，因为Dog、Cat都是Animal类型。

### 静态语言VS动态语言

* 对于静态语言，需要传入Animal类型，则传入的对象必须是Animal类型或者是Animal的子类，否则无法调用run方法。

* 对于Python动态语言，不一定传入Animal类型，只要保证传入的对象有一个run方法就行。

```python
class Timer(object):
    def run(self):
        print('Start...')
```

```python
run_twice(Timer())
```
