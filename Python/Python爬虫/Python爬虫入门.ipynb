### 读取文件并用BeautifulSoup加载

```python
from bs4 import BeautifulSoup  # 从bs4引入BeautifulSoup
# 读取文件内容到html变量里面
file_obj = open('douban.html', 'rb')  # 以读方式打开文件名为douban.html的文件
html = file_obj.read()  # 把文件的内容全部读取出来并赋值给html变量
file_obj.close()  # 关闭文件对象

soup = BeautifulSoup(html, 'lxml')  # 初始化BeautifulSoup
# print(soup)  # 输出BeautifulSoup转换后的内容
all_movies = soup.find('div',id="showing-soon")#先找到最大的div
# print(all_movies)#输出最大div里面的内容

list_name = []
list_href = []
list_date = []
list_type = []
list_area = []
list_lovers = []
list_img = []
for each_movie in all_movies.find_all('div',class_="item"):#从最大的div里面找到影片的div
#     print(each_movie)#输出每个影片div的内容
    all_a_tag = each_movie.find_all('a')#找到所有的a标签
#     print(all_a_tag)
    all_li_tag = each_movie.find_all('li')#找到所以得li标签
    
    all_img_tag = each_movie.find_all('img')
#     print(all_li_tag)
    movie_name = all_a_tag[1].text#从第二个a标签的文字内容提取影片内容
#     print(movie_name)
    movie_href = all_a_tag[1]['href']#从第二个a标签的文字内容提取影片链接
#     print(movie_href)
    movie_date = all_li_tag[0].text#影片日期
    movie_type = all_li_tag[1].text#影片类型
    movie_area = all_li_tag[2].text#影片地区
    movie_lovers = all_li_tag[3].text#影片喜欢人数
    movie_img = all_img_tag[0]['src']    
    
    list_name.append(movie_name)
    list_href.append(movie_href)
    list_date.append(movie_date)
    list_type.append(movie_type)
    list_area.append(movie_area)
    list_lovers.append(movie_lovers)
    list_img.append(movie_img)

#     print(all_img_tag)
    
    print('名字：{}，链接：{}，影片海报链接：{}，日期：{}，类型：{}，地区：{}，关注者：{}'.format(movie_name,movie_href,movie_img,movie_date,movie_type,movie_area,movie_lovers))
    
for i in range(len(list_name)):
#     print(i)
    file_obj = open('ok.txt','a',encoding="utf-8")
    file_obj.write('  [' + list_name[i]+ ']  ')
    file_obj.write('  [' + list_href[i]+ ']  ')
    file_obj.write('  [' + list_date[i]+ ']  ')
    file_obj.write('  [' + list_type[i]+ ']  ')
    file_obj.write('  [' + list_area[i]+ ']  ')
    file_obj.write('  [' + list_lovers[i]+ ']  ')
    file_obj.write('  [' + list_img[i]+ ']  ')
    
file_obj.close()    
```

### 数据保存到html

```python
#请求网页
url = "https://movie.douban.com/cinema/later/chengdu/"
response = requests.get(url)

 # 初始化BeautifulSoup方法一：利用网页字符串自带的编码信息解析网页
soup = BeautifulSoup(response.content.decode('utf-8'), 'lxml') 

# 初始化BeautifulSoup方法二：手动指定解析编码解析网页
# soup = BeautifulSoup(response.content, 'lxml', from_encoding='utf-8') 

# print(soup)  # 输出BeautifulSoup转换后的内容
all_movies = soup.find('div', id="showing-soon")  # 先找到最大的div
# print(all_movies)  # 输出最大的div的内容

html_file = open('data.html', 'w', encoding="utf-8")
html_file.write("""
<!DOCTYPE html>
<html>
<head>
    <meta charset="UTF-8">
    <title>豆瓣电影即将上映影片信息</title>
    <link href="https://cdn.bootcss.com/bootstrap/4.0.0/css/bootstrap.min.css" rel="stylesheet">
</head>
<body>
<h2 class="text-center">豆瓣电影即将上映影片信息</h2>
<table class="table table-striped table-hover mx-auto text-center">
    <thead>
        <tr>
            <th>影片名</th>
            <th>上映日期</th>
            <th>影片类型</th>
            <th>地区</th>
            <th>关注者数量</th>
        </tr>
    </thead>
    <tbody>
""")
for each_movie in all_movies.find_all('div', class_="item"):  # 从最大的div里面找到影片的div
    # print(each_movie)  # 输出每个影片div的内容
    all_a_tag = each_movie.find_all('a')
    all_li_tag = each_movie.find_all('li')
    movie_name = all_a_tag[1].text
    moive_href = all_a_tag[1]['href']
    movie_date = all_li_tag[0].text
    movie_type = all_li_tag[1].text
    movie_area = all_li_tag[2].text
    # 替换字符串里面的 想看 两个字为空，使得更加美观
    movie_lovers = all_li_tag[3].text.replace("想看", '')
    print('名字：{}，链接：{}，日期：{}，类型：{}，地区：{}， 关注者：{}'.format(
        movie_name, moive_href, movie_date, movie_type, movie_area, movie_lovers))
    html_file.write("""
        <tr>
            <td><a href="{}">{}</a></td>
            <td>{}</td>
            <td>{}</td>
            <td>{}</td>
            <td>{}</td>
        </tr>
    """.format(moive_href, movie_name, movie_date, movie_type, movie_area, movie_lovers))
html_file.write("""
     </tbody>
</table>
</body>
</html>
""")
html_file.close()
print("write_finished!")
```
